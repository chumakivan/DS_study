# 22. Сколько суммарно вышло фильмов летом? (за июнь, июль, август)

```python
answers['22'] = '2. 450'
```

```python
def time_of_year(date_str):
    date_month = date_str.split('/')[0]
    if date_month in '345':
        return 'spring'
    elif date_month in '678':
        return 'summer'
    elif date_month in ('9','10','11'):
        return 'fall'
    else:
        return 'winter'
```

```python
data['time_of_year'] = data.release_date.apply(time_of_year)
```

```python
data.groupby(by='time_of_year').count()
```
