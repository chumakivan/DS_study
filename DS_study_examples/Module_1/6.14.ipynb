```python
df = pd.read_csv('data_sf.csv.xls')
```

С помощью сводной таблицы и функции loc посчитайте, сколько получают ("Wage") русские футболисты ("Russia"), играющие за ФК "AS Monaco".

```python
pivot = df.pivot_table(values='Wage',columns='Club',index='Nationality',aggfunc = 'sum',fill_value=0)
```

```python
pivot.loc['Russia']['AS Monaco']
```
